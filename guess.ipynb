## guess the number 

```python
import random

def guess_number_game():
    """Main game function for Guess the Number with difficulty levels"""
    
    print("=" * 50)
    print("Welcome to Guess the Number Game!")
    print("=" * 50)
    
    # Difficulty levels with their attempt limits
    difficulty_levels = {
        '1': {'name': 'Easy', 'attempts': 10, 'range_min': 1, 'range_max': 100},
        '2': {'name': 'Medium', 'attempts': 7, 'range_min': 1, 'range_max': 100},
        '3': {'name': 'Hard', 'attempts': 5, 'range_min': 1, 'range_max': 100}
    }
    
    # Display difficulty options
    print("\nChoose Difficulty Level:")
    print("-" * 35)
    print("1. Easy   (10 attempts)")
    print("2. Medium (7 attempts)")
    print("3. Hard   (5 attempts)")
    print("-" * 35)
    
    # Get valid difficulty choice
    while True:
        choice = input("\nEnter your choice (1/2/3): ").strip()
        if choice in difficulty_levels:
            break
        print("Invalid choice! Please enter 1, 2, or 3.")
    
    # Get game settings for chosen difficulty
    settings = difficulty_levels[choice]
    max_attempts = settings['attempts']
    min_num = settings['range_min']
    max_num = settings['range_max']
    difficulty_name = settings['name']
    
    # Generate random number
    secret_number = random.randint(min_num, max_num)
    attempts_used = 0
    attempts_left = max_attempts
    
    print(f"\nGame Started! ({difficulty_name} Mode)")
    print(f"I'm thinking of a number between {min_num} and {max_num}")
    print(f"You have {max_attempts} attempts to guess it!")
    print("-" * 50)
    
    # Main game loop
    while attempts_used < max_attempts:
        attempts_left = max_attempts - attempts_used
        
        # Get user's guess with validation
        try:
            guess = int(input(f"\nAttempt {attempts_used + 1}/{max_attempts} (Remaining: {attempts_left}): "))
        except ValueError:
            print("Please enter a valid number!")
            continue
        
        # Check if guess is within range
        if guess < min_num or guess > max_num:
            print(f"Please guess between {min_num} and {max_num}!")
            continue
        
        attempts_used += 1
        
        # Compare guess with secret number
        if guess == secret_number:
            print(f"\nCORRECT!")
            print(f"You guessed the number {secret_number} in {attempts_used} attempts!")
            print(f"Difficulty: {difficulty_name} Mode")
            break
        elif guess < secret_number:
            print("Too Low! Try a higher number.")
        else:
            print("Too High! Try a lower number.")
        
        # Show remaining attempts
        if attempts_used < max_attempts:
            print(f"You have {max_attempts - attempts_used} attempts left.")
    
    # Game over - out of attempts
    if attempts_used == max_attempts and guess != secret_number:
        print("\nGAME OVER!")
        print(f"You've used all {max_attempts} attempts!")
        print(f"The secret number was: {secret_number}")
        print(f"Difficulty: {difficulty_name} Mode")
    
    # Ask for replay
    print("\n" + "=" * 50)
    play_again = input("Would you like to play again? (yes/no): ").strip().lower()
    if play_again in ['yes', 'y']:
        print("\n" * 2)
        guess_number_game()
    else:
        print("\nThanks for playing! Goodbye!")

# Run the game
guess_number_game()
```
